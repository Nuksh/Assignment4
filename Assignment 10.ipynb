```python
reader_books = pd.pivot_table(book_df,index='User.ID',
                           columns='Book.Title',
                           values='Book.Rating').reset_index(drop=True)
```

```python
reader_books
```

```python
reader_books.index = book_df.User_id.unique()
```

```python
reader_books
```

```python
reader_books.fillna(0, inplace=True)
```

```python
reader_books
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
reader_sim = 1 - pairwise_distances( reader_books.values,metric='cosine')
```

```python
reader_sim
```

```python
reader_sim_df = pd.DataFrame(reader_sim)
```

```python
reader_sim_df.index = book_df['User_id'].unique()
reader_sim_df.columns = book_df['User_id'].unique()
```
